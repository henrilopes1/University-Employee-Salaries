```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

data = pd.read_csv('/content/higher_ed_employee_salaries.csv')

# 1. Como os ganhos variam entre diferentes instituies?
earnings_by_school = data.groupby('School')['Earnings'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 8))
earnings_by_school.head(10).plot(kind='bar', color='teal')
plt.title('Average Earnings by Institution (Top 10)')
plt.xlabel('Institution')
plt.ylabel('Average Earnings')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 2. Existe uma tendncia nos ganhos dos funcionrios ao longo dos anos?
earnings_over_years = data.groupby('Year')['Earnings'].mean()
plt.figure(figsize=(10, 6))
earnings_over_years.plot(kind='line', marker='o', color='purple')
plt.title('Trend of Average Earnings Over Years')
plt.xlabel('Year')
plt.ylabel('Average Earnings')
plt.grid(True)
plt.show()
```

```python
# 3. Quais so as diferenas salariais entre diferentes categorias de emprego?
earnings_by_job = data.groupby('Job Description')['Earnings'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 8))
earnings_by_job.head(10).plot(kind='bar', color='orange')
plt.title('Average Earnings by Job Description (Top 10)')
plt.xlabel('Job Description')
plt.ylabel('Average Earnings')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 4. Como a distribuio de salrios varia entre diferentes tipos de cargos na "University of Akron"?
# Filtrar dados para 'University of Akron' e selecionar os cargos mais frequentes
filtered_data = data[data['School'] == 'University of Akron']
top_jobs = filtered_data['Job Description'].value_counts().nlargest(5).index

# Criar um grfico para cada tipo de cargo
plt.figure(figsize=(12, 8))
for job in top_jobs:
    subset = filtered_data[filtered_data['Job Description'] == job]['Earnings'].dropna()
    subset.plot(kind='density', label=job)

plt.title('Salary Distribution by Job Type at University of Akron')
plt.xlabel('Earnings')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
# 5. Quais departamentos tm os maiores e menores ganhos mdios?
earnings_by_department = data.groupby('Department')['Earnings'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 8))
earnings_by_department.head(10).plot(kind='bar', color='blue')
plt.title('Top 10 Highest Earning Departments')
plt.xlabel('Department')
plt.ylabel('Average Earnings')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 6. Como a variao salarial se compara entre departamentos dentro das mesmas instituies?
# Vamos visualizar para uma escola especfica
school_department_earnings = data[data['School'] == 'University of Akron'].groupby('Department')['Earnings'].std()
plt.figure(figsize=(12, 8))
school_department_earnings.dropna().sort_values(ascending=False).head(10).plot(kind='bar', color='red')
plt.title('Variability of Earnings by Department at University of Akron')
plt.xlabel('Department')
plt.ylabel('Standard Deviation of Earnings')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 7. Existe correlao entre a antiguidade no cargo (baseada no ano) e os ganhos?
plt.figure(figsize=(10, 6))
plt.scatter(data['Year'], data['Earnings'], alpha=0.5, color='magenta')
plt.title('Correlation Between Year and Earnings')
plt.xlabel('Year')
plt.ylabel('Earnings')
plt.grid(True)
plt.show()
```
